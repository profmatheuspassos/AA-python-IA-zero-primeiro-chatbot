# Aula 7: Dando accesso a sites para o nosso bot

O nosso objetivo é construir um bot que tenha acesso a informações personalizadas, que podem estar distribuídas em diversas fontes, como: sites, vídeos de Youtube, pdfs, etc.

A partir de agora, vamos dar opções ao nosso bot, para que ele consiga acessar as informações externas que ele utilizará como base para a geração de respostas mais específicas.

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://dponapratica.com.br/2023/05/a-nova-abnt-nbr-iso-iec-27005-2023/') # O loader do LangChain sempre retorna uma lista

lista_documentos = loader.load()
```

```python
documento = ''
for doc in lista_documentos:
    documento += doc.page_content
```

```python
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

_ = load_dotenv(find_dotenv())

chat = ChatGroq(model='llama-3.1-70b-versatile')
```

```python
template = ChatPromptTemplate.from_messages({
    ('system', 'Você é um assistente amigável chamado Asimo e tem acesso às seguintes informações para dar as suas respostas: {documentos_informados}'),
    ('user', '{input}')
})

chain = template | chat

resposta = chain.invoke({'documentos_informados': documento, 'input': 'Quais as principais características da ISO 27005?'})
```

```python
print(resposta.content)
```

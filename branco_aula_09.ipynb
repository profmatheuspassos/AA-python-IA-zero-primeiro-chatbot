# Aula 9: Criando a estrutura final do nosso ChatBot

A nossa ideia agora é criar uma estrutura onde o usuário possa dizer se ele quer capturar informações do Youtube, de um pdf ou de um site e, a partir desse direcionamento, criamos documentos diferentes para o modelo de linguagem utilizar na criação da reposta.

```python
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

_ = load_dotenv(find_dotenv())

chat = ChatGroq(model='llama-3.1-70b-versatile')

def resposta_bot(mensagens):
    mensagens_modelo = [('system', 'Você é um modelo amigável chamado Asimo.')]
    mensagens_modelo += mensagens
    template = ChatPromptTemplate.from_messages(mensagens_modelo)
    chain = template | chat
    return chain.invoke({}).content
```

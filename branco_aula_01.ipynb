## Variáveis

**Uma variável é uma forma de armazenar dados na memória do computador para que possam ser utilizados e manipulados ao longo do código.** Ela age como um "rótulo" que aponta para um valor ou conjunto de dados.


### O que são variáveis?

**Variáveis são as caixas...**

Imagine que você tem **uma caixa** em sua casa, e essa caixa tem **um rótulo**. **O rótulo diz o que tem dentro da caixa**. Por exemplo, se você colocar "Doces" no rótulo e dentro da caixa colocar balas, chocolates, etc., toda vez que você quiser doces, sabe que é só olhar na caixa que tem esse rótulo.

Na programação, uma variável é como essa caixa com um rótulo. Você dá um **nome à variável** (como o rótulo da caixa) e coloca **um valor dentro dela** (como os doces na caixa). Então, sempre que você precisar desse valor, pode usar o nome da variável para encontrá-lo.

Por exemplo:

- Nome da caixa: **nome**
- O que tem dentro da caixa: **"Adriano"**
Então, sempre que você quiser saber o nome, basta olhar para a variável nome, que está guardando o valor "Adriano".

Dessa forma, as variáveis ajudam a organizar e lembrar de informações no seu código, como as caixas ajudam você a organizar seus objetos em casa.

```python
from IPython.display import YouTubeVideo
YouTubeVideo('8wJCm2MN0YU', width=800, height=600)
```

**A memória do computador é a estante...**

A memória do computador é como a prateleira onde as caixas são armazenadas. Quando você cria uma variável no seu programa, é como se você pegasse uma caixa vazia na prateleira e colasse um rótulo nela. Esse rótulo permite que você saiba o que está guardado dentro da caixa sem precisar abri-la toda vez.

Por exemplo, ao criar a variável nome idade e armazena o valor "Adriano" nela, é como se você colocasse uma caixa com o rótulo nome na prateleira, e dentro dessa caixa, o valor "Adriano" está guardado. Sempre que o programa precisar saber o nome, ele vai até a prateleira, encontra a caixa com o rótulo nome, e pega o valor que está dentro.

Na memória do computador, essas caixas são locais onde os dados são armazenados temporariamente enquanto o programa está sendo executado. O rótulo (nome da variável) facilita para o computador encontrar e usar os dados sem precisar lembrar de onde exatamente eles estão armazenados na prateleira (memória).

Assim como na analogia da prateleira cheia de caixas, o computador organiza e recupera informações de forma eficiente usando essas "caixas" (variáveis) para manter tudo em ordem.

```python
nome_pessoa = "Fernando"
idade_pessoa = 28
altura_pessoa = 1.76
```

## Tipos de Dados

Tipos de dados são as diferentes categorias que os valores podem ter em um programa. Cada tipo de dado define o que pode ser feito com aquele valor e como ele é armazenado na memória do computador.

### Principais tipos em Python
- *Números Inteiros (int)*: São números sem casas decimais, como 5, -2, ou 1000.

- *Números Decimais (float)*: São números que têm uma parte decimal, como 3.14, -0.5, ou 100.75. **Atenção! Por padrão, Python utiliza a notação em inglês, portanto, a casa decimal DEVE ser representada por ponto (.) e não por vírgula (,)!**

- *Textos (str)*: Conhecidos como strings, são sequências de caracteres, como palavras ou frases, por exemplo, "Olá, Chatbot!" ou "Python".

- *Valores Lógicos (bool)*: São valores que só podem ser True (verdadeiro) ou False (falso). Eles são usados para tomar decisões no código, como verificar se algo é verdadeiro ou falso.

Por que é importante?
Os tipos de dados são importantes porque o computador precisa saber que tipo de valor ele está lidando para realizar operações corretamente. Por exemplo, você pode somar dois números, mas não pode somar uma palavra com um número, pois são tipos de dados diferentes.


```python
nome_pessoa = "Fernando" # string
idade_pessoa = 28 # int
altura_pessoa = 1.76 # float
```

### Operadores Aritméticos

Os operadores aritméticos são usados para realizar operações matemáticas básicas com números. Aqui estão os principais operadores aritméticos e o que eles fazem:

- \+ (Soma): Adiciona dois valores.

- \- (Subtração): Subtrai um valor de outro.

- \* (Multiplicação): Multiplica dois valores.

- / (Divisão): Divide um valor por outro.

- ** (Exponenciação): Eleva um valor à potência de outro.


```python
print(idade_pessoa ** altura_pessoa)
```

### Entrada de dados com a função *input*

Imagine que você está em uma loja e o atendente pergunta: "Qual é o seu nome?" Ele faz uma pausa, esperando que você responda. Quando você diz seu nome, o atendente anota essa informação e a usa para continuar o atendimento.

Na programação, a função input é como esse atendente fazendo uma pergunta. O programa faz uma pausa, esperando que o usuário digite uma resposta. Quando o usuário digita algo, o programa "anota" essa resposta e pode usá-la para realizar alguma ação ou continuar o processo.

```python
idade = input("Qual a sua idade? ")
print(idade)
```

## Formatação com f-string

A formatação com f-string em Python é uma maneira simples e eficiente de inserir valores de variáveis dentro de uma string. Para usar, você coloca um f antes das aspas que delimitam a string e insere as variáveis dentro de chaves {} diretamente na string.

```python
print(f"O nome é {nome_pessoa}, que tem {idade_pessoa} e como altura, {altura_pessoa}m.")
```

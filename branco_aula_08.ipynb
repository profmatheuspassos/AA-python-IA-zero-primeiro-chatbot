# Aula 8: Acessando vídeos de Youtube e PDFs

```python
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

_ = load_dotenv(find_dotenv())

chat = ChatGroq(model='llama-3.1-70b-versatile')
```

## Acessando o Youtube

E agora vamos mostrar o quão fácil é de acessar vídeos de youtube utilizando LangChain!

```python
from langchain_community.document_loaders import YoutubeLoader

url = 'https://www.youtube.com/watch?v=1gOMpczyTpI&t=4s'

loader_yt = YoutubeLoader.from_youtube_url(
    url,
    language=['pt']
)

lista_documentos_yt = loader_yt.load()
```

```python
documento_yt = ''
for doc in lista_documentos_yt:
    documento_yt += doc.page_content
```

```python
template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente amigável que possui as seguintes informações para formular uma resposta: {informacoes}'),
    ('user', '{input}')
])

chain_youtube = template | chat

resposta_yt = chain_youtube.invoke({'informacoes': documento_yt, 'input': 'Qual o título do vídeo e qual seu tema central?'})

print(resposta_yt.content)
```

## Acessando PDFs

```python
from langchain_community.document_loaders import PyPDFLoader

caminho = 'arquivos/2024-07-15_Discussion_Paper_Hamburg_DPA_KI_Models.pdf'

loader_pdf = PyPDFLoader(caminho)

lista_documentos_pdf = loader_pdf.load()
```

```python
documento_pdf = ''

for doc in lista_documentos_pdf:
    documento_pdf += doc.page_content
```

```python
template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente amigável que possui as seguintes informações para formular uma resposta: {informacoes}'),
    ('user', '{input}')
])

chain_pdf = template | chat

resposta_pdf = chain_pdf.invoke({'informacoes': documento_pdf, 'input': 'Quais as principais conclusões do documento? Apresente sua resposta em plain text, sem markdown.'})

print(resposta_pdf.content)
```
